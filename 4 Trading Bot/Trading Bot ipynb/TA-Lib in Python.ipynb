```
import talib
from talib import abstract
import numpy as np
import yfinance as yf

ticker = input("Enter Ticker: ")
# ticker = 'MSFT'

data = yf.download(ticker, start="2020-01-01", end="2020-08-01")
data = data[['Open', 'High', 'Low', 'Adj Close', 'Volume']]
data.rename(columns={'Open':'open', 'High':'high', 'Low':'low', 'Adj Close':'close', 'Volume':'volume'}, inplace=True)

data['RSI'] = abstract.RSI(data)
print(data)


import matplotlib.pyplot as plt

plt.plot(data.index,data['close'])
plt.plot(data.index,data[data.columns[-1]])
plt.title("Analysis of "+ticker) 
plt.xlabel("Date") 
plt.ylabel("Price") 
plt.show()
```
